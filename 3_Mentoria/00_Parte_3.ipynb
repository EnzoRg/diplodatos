```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.dates as mdates
from scipy.stats import boxcox
```

```python
plt.style.use('ggplot')
color = sns.color_palette().as_hex()
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv('../Data/output_eda.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df, x=df.index, y='traffic_flow', label='Trfico de vehculos', color=color[1])
plt.title('Trfico de vehculos 2019-2024', fontsize=14)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.ylabel('Cantidad de vehculos')
plt.xlabel('Tiempo')
plt.legend(loc='upper left')
plt.show()
```

```python
df_2 = boxcox(df.traffic_flow)
```
